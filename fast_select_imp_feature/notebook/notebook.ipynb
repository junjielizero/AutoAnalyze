# 可视化

```python
iv_idx = iv_df.index

# 解读

"""
红线代表整体数据生存率，高于红线，说明该区间生存率相对较高，反之较低，如女性生存率高于男性
下图按照由大到小影响程度排序，可见性别离开红线差异较大，年龄较低
"""

for i in iv_idx:
    plt.figure(figsize=(10,5));
    plt.title(i)
    plt.bar(bins[i]['bin'].values,bins[i]['badprob'].values)
    plt.axhline(y=df['target'].mean(), color='r', linestyle='-')
```
